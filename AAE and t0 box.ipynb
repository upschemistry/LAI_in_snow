```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
```

```python
fig, ax = plt.subplots(figsize=(6,4))
plt.xlim([3.5, 8])
plt.xlabel("$t_0 (s)$")
plt.ylim([0.6, 2.8])
plt.ylabel("Å")
plt.grid(True)

rect = Rectangle((5, 1.5), 2, 1, fill=False, linewidth=2); ax.add_patch(rect)

rect = Rectangle((6.3, 0.7), 0.4, 0.5, fill=False, linewidth=2); ax.add_patch(rect)

rect = Rectangle((4.2, 1.0), 0.4, 0.2, fill=False, linewidth=2); ax.add_patch(rect)
```
