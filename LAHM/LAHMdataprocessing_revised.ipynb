Code to process LAHM data imported as a .csv file

```python
#generic resources, not all necessarily needed for this
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
import csv
import pandas as pd
```

```python
foldername = 'Rainier (UPS UPS UPS)/'
csvfilename = foldername+'filelist_all_2026.csv'
print(csvfilename)
data = pd.read_csv(csvfilename)
print(data.columns)

plt.figure()
x = data['t0']
y = data['ngperg']
plt.scatter(x,y)
plt.xlabel('t$_0$ (s)')
plt.ylabel('LAI concentration (ppb eBC)')
plt.yscale('log')
```
