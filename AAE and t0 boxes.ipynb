```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
import numpy as np
```

```python
fig, ax = plt.subplots(figsize=(6,4))
plt.xlim([3.5, 8])
plt.xlabel("$t_0 (s)$")
plt.ylim([-1.2, 3.2])
plt.ylabel("Å")
plt.grid(True)
alpha = 0.5

rect = Rectangle((3.95, 1.5), 2.5, 1.5, fill=True, color='red', alpha=alpha/2); ax.add_patch(rect)
plt.text(4.7, 2.1, 'igneous rock')

rect = Rectangle((6.3, 1.5), 0.4, 0.6, fill=True, color='orange', alpha=1); ax.add_patch(rect)
plt.text(6.2,1.3,'pine pollen')

rect = Rectangle((4.2, 0.5), 0.3, 0.8, fill=True, color='black', alpha=alpha); ax.add_patch(rect)
plt.text(3.9,0.3,'Rainier black')

rect = Rectangle((5.1, -1.0), 0.6, 1, fill=True, color='black', alpha=alpha/2); ax.add_patch(rect)
plt.text(5.8,-.5,'fullerene standard')

MtRainierMuir_AAE = np.array([1.8, 1.4])
MtRainierMuir_t0 = np.array([7.1, 5.4]) 
plt.plot(MtRainierMuir_t0,MtRainierMuir_AAE,'o',label='Rainier (Muir)')

MtRainierSummit_AAE = np.array([.24, 1.5])
MtRainierSummit_t0 = np.array([5.4, 6.3])
plt.plot(MtRainierSummit_t0,MtRainierSummit_AAE,'x',label='Rainier (summit)')

ax.legend(bbox_to_anchor=(1.2, 0.5))
```
