```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
Nuc_index = np.array([4,5,6,7])
L = np.array([.75,.86,1.03,1.81])

p = np.polyfit(Nuc_index,L,3)
Nuc_index_theory = np.linspace(np.min(Nuc_index),11)

plt.figure()
plt.plot(Nuc_index,L,'o')
plt.plot(Nuc_index_theory,np.polyval(p,Nuc_index_theory))
plt.grid(True)
plt.xlabel('Nuc index')
plt.ylabel('L (g/cm^2)')
```
